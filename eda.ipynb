```python
import pandas as pd

df = pd.read_csv("netflix_titles.csv")

# Basic info
df.info()

# Handle missing values
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)

# Drop rows where important fields missing
df.dropna(subset=['date_added'], inplace=True)

# Convert date
df['date_added'] = df['date_added'].str.strip()
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['date_added'] = pd.to_datetime(df['date_added'])

df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='type', data=df)
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

```python
df['year_added'] = df['date_added'].dt.year
```

```python
df['year_added'].value_counts().sort_index().plot(kind='line')
plt.title("Content Added Over Years")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar')
plt.title("Top Content Producing Countries")
plt.show()
```

```python
df['listed_in'].value_counts().head(10).plot(kind='bar')
plt.title("Top Genres")
plt.show()
```

Netflix has significantly more Movies than TV Shows
Content addition increased rapidly after 2015
United States produces the highest content
Drama and International content dominate the platform
Recent years show higher diversity in genres
